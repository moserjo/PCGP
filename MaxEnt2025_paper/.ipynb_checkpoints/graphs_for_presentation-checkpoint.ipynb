# Bipendulum animation

```python


# --- Parameters ---
g = 9.81   # gravity
L1, L2 = 1.0, 2.0   # rod lengths
m1, m2 = 1.0, 1.0   # masses
dt = 0.02           # timestep
T = 10              # total simulation time
nsteps = int(T / dt)

# initial angles and angular velocities
theta1, theta2 = np.pi/2, np.pi/2
omega1, omega2 = 0.0, 0.0

# storage
t_vals = []
th1_vals, th2_vals = [], []



def ex2_analytic_solution(x, l1 = 1, l2 = 2., g = 9.81, noise = 0):
    X = 10
    a1 = math.sqrt(g/l1)
    a2 = math.sqrt(g/l2)
    theta_1 = 3*np.sin(a1*x) + X/l1/(a1**2 - 4)*np.sin(2*x) + noise*np.random.randn(x.shape[0])
    theta_2 = -5*np.cos(a2*x)  + X/l2/(a2**2 - 4)*np.sin(2*x)+ noise*np.random.randn(x.shape[0])
    u = X*np.sin(2*x)+ noise*np.random.randn(x.shape[0])
    #y = np.stack([f1, f2, u1], axis = -1)
    return theta_1/15, theta_2/15, u/15
nsteps =50
T = np.linspace(0, 6, nsteps)
theta_1, theta_2, u = ex2_analytic_solution(T)

# --- Coordinates ---
x1 = L1 * np.sin(th1_vals)
y1 = -L1 * np.cos(th1_vals)
x2 = x1 + L2 * np.sin(th2_vals)
y2 = y1 - L2 * np.cos(th2_vals)

l_bar = 2
bar_x = np.stack([u - l_bar/2, u+l_bar/2], axis = -1)


x1 = bar_x[:,0] + np.sin(theta_1)
x2 = bar_x[:,1] + np.sin(theta_2)
y1 = -np.cos(theta_1)
y2 = -np.cos(theta_2)

# --- Animation loop (saving frames) ---
for i in range(0, nsteps):   # skip some frames for speed
    fig, ax1= plt.subplots(1, 1, figsize=(8,4))

    # ---- subplot 1: double pendulum ----
    ax1.set_xlim(-2, 5)
    ax1.set_ylim(-2.5, 0.5)
    ax1.set_aspect('equal')
    ax1.axis('off')

    # top bar
    ax1.text(u[i], 0.1, "u")
    ax1.plot([bar_x[i,0] ,bar_x[i,1]], [0.0, 0.0], color='steelblue', lw=4)

    # rods
    ax1.plot([bar_x[i,0], L1*x1[i]], [0, L1*y1[i]], color='k', lw=2)
    ax1.plot([bar_x[i,1], L2*x2[i]], [0, L2*y2[i]], color='k', lw=2)

    # masses
    ax1.add_patch(Circle((L1*x1[i], L1*y1[i]), 0.05, fc='orange', zorder=3))
    ax1.add_patch(Circle((L2*x2[i], L2*y2[i]), 0.05, fc='red', zorder=3))

    # ---- subplot 2: example signal (angles vs time) ----
    #ax2.plot(t_vals[:i], th1_vals[:i], label=r'$\theta_1$')
    #ax2.plot(t_vals[:i], th2_vals[:i], label=r'$\theta_2$')
    #ax2.set_xlim(0, T)
    #ax2.set_ylim(-np.pi, np.pi)
    #ax2.set_xlabel('time [s]')
    #ax2.legend()

    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, f"new_bipendulum_{i:04d}.png"), dpi=150)
    plt.show()
    plt.close(fig)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm
from scipy.optimize import minimize_scalar
from scipy.stats import norm
# --- Parameters for skew-normal ---
xi = 0       # location
omega = 1    # scale
alpha = 3    # positive = right skew

# --- Sample from the true distribution ---
N = 10000
samples = skewnorm.rvs(a=alpha, loc=xi, scale=omega, size=N)


neg_logpdf = lambda x: -skewnorm.logpdf(x, a=alpha, loc=xi, scale=omega)
res = minimize_scalar(neg_logpdf)
mode = res.x

# --- Second derivative (curvature) at the mode ---
eps = 1e-5
f = lambda x: skewnorm.logpdf(x, a=alpha, loc=xi, scale=omega)
fpp = (f(mode+eps) - 2*f(mode) + f(mode-eps)) / eps**2  # finite difference
sigma_laplace = np.sqrt(-1 / fpp)
# --- Plot ---
x = np.linspace(min(samples), max(samples), 500)

plt.figure()
plt.plot(x, skewnorm.pdf(x, a=alpha, loc=xi, scale=omega),
         'b-', lw=2, label="True")
plt.legend(["True",])
plt.savefig(os.path.join(save_dir, f"laplace_0.png"), dpi=300)
plt.figure()
plt.plot(x, skewnorm.pdf(x, a=alpha, loc=xi, scale=omega),
         'b-', lw=2, label="True")
plt.plot(x, norm.pdf(x, loc=mode, scale=sigma_laplace),
         'r--', lw=2, label="Laplace approx")
plt.legend(["True","Laplace"])
plt.savefig(os.path.join(save_dir, f"laplace_1.png"), dpi=300)
plt.figure()
plt.plot(x, skewnorm.pdf(x, a=alpha, loc=xi, scale=omega),
         'b-', lw=2, label="True")
plt.plot(x, norm.pdf(x, loc=mode, scale=sigma_laplace),
         'r--', lw=2, label="Laplace approx")
plt.hist(samples, bins=50, density=True, alpha=0.5, label="_MCMC")
plt.legend(["True","Laplace", "MCMC"])

plt.savefig(os.path.join(save_dir, f"laplace_2.png"), dpi=300)



plt.show()
```

```python
#get data
data_AE = np.load(os.path.join(data_dir_ex2, "Ex2_AE_with_n234.npz"))
data_DKL = np.load(os.path.join(data_dir_ex2, "Ex2_DKL_with_n234.npz"))
sigma_list = data_DKL["sigma_list"]
n_list = list(data_DKL["n_list"])
all_AE_end2 = data_AE["all_AE_end2"]
all_AE_end6 = data_AE["all_AE_end6"]
all_DKL_end2 = data_DKL["all_DKL_end2"]#np.where(data_DKL["all_DKL_end2"]<=2e1, data_DKL["all_DKL_end2"], 1e2) #Ausreißer weggeben
all_DKL_end6 = data_DKL["all_DKL_end6"]#np.where(data_DKL["all_DKL_end6"]<=5e1, data_DKL["all_DKL_end6"], 1e2) #Ausreißer weggeben
n_list.remove(4)
#for plotting
width = 0.05
group_spacing = 0.3
num_boxes_per_group = 4
scale_factor = 40
base_size = 20
symbol = ["^", "^", "s", "s"]
markers = ["x", "+", 1/np.sqrt(2), 1]
colors = list(mcolors.TABLEAU_COLORS.values())

fig = plt.figure(figsize=(18, 6))
gs = gridspec.GridSpec(1, 2, width_ratios=[2, 2], wspace=0.3)

# --- LEFT AXIS: split into 2 rows ---
gs_left = gridspec.GridSpecFromSubplotSpec(2, 1, subplot_spec=gs[0, 0], hspace=0.05)
ax00 = fig.add_subplot(gs_left[0, 0])  # top subplot
ax01 = fig.add_subplot(gs_left[1, 0], sharey=ax00)  # bottom subplot

# --- RIGHT AXIS ---
ax1 = fig.add_subplot(gs[0, 1])

j_list = [0, 1, 2, 3]

# --- LEFT TOP: j=0,1 ---
for n_idx, n in enumerate(n_list):
    if n != 50: 
        n_idx+=1
        continue
    group_pos = n_idx * (num_boxes_per_group * width + group_spacing)
    for j in range(2):  # j=0,1
        pos = group_pos + 2 * j * width
        for sigma_idx, sigma in enumerate(sigma_list):
            if sigma_idx != 0: 
                sigma_idx+=1
                continue
            data = all_AE_end6[n_idx, sigma_idx, j]
            x_vals = np.full(len(data), pos)
            marker_size = base_size + scale_factor * sigma_idx
            ax00.scatter(x_vals, data,
                         marker=symbol[j], color=colors[j % len(colors)],
                         alpha=0.1, s=marker_size,
                         label=f'σ={sigma:.2f}' if n_idx == 0 and j == 0 else None)
            """if n in [2,3,4,5, 7]:
                extra_y = np.atleast_1d(all_AE_end2[n_idx, sigma_idx, j])
                ax00.scatter(np.full(len(extra_y), pos), extra_y,
                            marker=markers[j], color="black", alpha=0.1,#colors[j % len(colors)]
                            s=marker_size*markers[j+2])"""
            ax00.tick_params(labelbottom=False, bottom = False, labelsize=20)
ax00.text(0, 1e-7, r"""$\Delta \ell_1$""", fontsize=24)
# --- LEFT BOTTOM: j=2,3 ---
for n_idx, n in enumerate(n_list):
    print(n_idx)
    if n != 50: 
        n_idx+=1
        continue
        
    group_pos = n_idx * (num_boxes_per_group * width + group_spacing)
    print(group_pos)
    for j in range(2, 4):  # j=2,3
        pos = group_pos + 2 * (j-2) * width  # reset spacing for bottom row
        for sigma_idx, sigma in enumerate(sigma_list):
            if sigma_idx != 0: 
                sigma_idx+=1
                continue
            data = all_AE_end6[n_idx, sigma_idx, j]
            x_vals = np.full(len(data), pos)
            marker_size = base_size + scale_factor * sigma_idx
            ax01.scatter(x_vals, data,
                         marker=symbol[j], color=colors[(j-2) % len(colors)],
                         alpha=0.2, s=marker_size,
                         label=f'σ={sigma:.2f}' if n_idx == 0 and j == 2 else None)
            """if n in [2,3,4,5, 7]:
                extra_y = np.atleast_1d(all_AE_end2[n_idx, sigma_idx, j])
                ax01.scatter(np.full(len(extra_y), pos), extra_y,
                            marker=markers[j-2], color = "black" , alpha=0.1,#color=colors[(j-2) % len(colors)]
                            s=marker_size*markers[j])"""
ax01.text(0, 1e-7, r"""$\Delta \ell_2$""", fontsize=24)
# --- format LEFT subplots ---
ax00.set_xlim(-num_boxes_per_group * width/2, 3 + num_boxes_per_group * width)
ax01.set_xlim(-num_boxes_per_group * width/2, 3 + num_boxes_per_group * width)
ax00.set_ylim(2e-8, 9e0)
ax01.set_ylim(2e-8, 9e0)
for ax in [ax00, ax01]:
    AE_xtick_positions = [i * (num_boxes_per_group * width + group_spacing)
                          + width*num_boxes_per_group/2 for i in range(len(n_list))]
    AE_xtick_labels = ["n = 2      " if n == 2 else f"{n}" for n in n_list]
    ax.set_xticks(AE_xtick_positions, AE_xtick_labels)
    ax.set_yscale("log")
    ax.grid(True, axis='y', linestyle='--', alpha=0.5)
ax00.set_title("Absolute MAP errors")
#ax01.set_title("Absolute MAP errors (j=2,3)")
#ax01.set_ylabel("Absolute Error")
ax00.legend(handles = [Patch(facecolor=colors[0], label = "PIGP"), Patch(facecolor=colors[1], label = "NSB") ], loc = 8)
ax1.axis("off")
plt.savefig(os.path.join(save_dir, "Ex2_metrics_0.png"), dpi=300,bbox_inches='tight' )
```
